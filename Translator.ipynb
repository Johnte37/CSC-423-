```python
import csv
```

```python
def read_csv_file(file_path):
    """Reads a CSV file and returns a dictionary"""
    with open(file_path, 'r', encoding='utf-8-sig') as file:
        reader = csv.reader(file)
        # Skip the header row
        next(reader)
        data = {row[0]: row[1] for row in reader}
    return data
```

```python
dictionary_file_path = r"C:\Users\User\Downloads\b8tc96xf3h-1\Common Swahili Slangs.csv"
dictionary = read_csv_file(dictionary_file_path)
```

```python
def translate_word(word):
    """Translates a Sheng word to Swahili"""
    if word in dictionary:
        return dictionary[word]
    else:
        return "Translation not found"
```

```python
print("Sheng Words:")
for sheng_word in dictionary.keys():
    print(sheng_word)
```

```python
sheng_word = input("Enter a Sheng word: ")
translation = translate_word(sheng_word.lower())
print("Translation: ", translation)
```
